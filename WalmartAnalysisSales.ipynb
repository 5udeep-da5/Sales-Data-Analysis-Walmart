# Initial Insights

The dtaset we are analysis is of walmart. This dataset consists of information regarding the sales data of Walmart for the year 2019. We have made some preliminary analysis of the dataset and drawn some initial insights which are as follows.
 1. The dataset consists of 1000 rows and 17 columns. The dataset is arranged with respect of the invoice id of the sales.
 2. Since the column of invoice_id is not necessary for our analysis, it is dropped.
 3. All the columns appear to be of correct data type except the date and time column which is in object datatype and moving forward feature engineering needs to be applied to ectract addtional insights.
 4. There are no null values or duplicate values in the dataset so there is no need for null value treatment.
 5. The highest invoice generating branch is branch is Branch A which generated 340 invoices followed by Branch B generating 332 invoices and then Branch C which generated 328 invoices.
 6. Those customer who are enrolled in any form of loyalty program with the store have generated more invoices i.e. 501 and non members have generated 499 invoices. This needs to be further analysed in terms of value.
 7. The product line which was most bought was Fashion Accessories followed by Food and Beverages and Electronic accessories.
 8. The most preferred mode of payment is Ewallet followed by Cash and Credit card.
 9. The agerage unit price of the products is 55.6 with minimum price being 26.49 and maximum price being 99.96.
 10. The average quantity bought by customers is 5.51 with minimum being 1 and maximum quantity bought is 10.

```python
# Feature engineering

#converting date column which is in object type into datetime type and extract year, month and day
```

```python
df.head()
```

```python
df.Date= pd.to_datetime(df.Date)
df.Date
```

```python
#Year month and day
```

```python
df.Date.dt.year
```

```python
df.Date.dt.month
```

```python
df.Date.dt.day
```

```python
df["Year"]= df.Date.dt.year
df["Month"]= df.Date.dt.month
df["Day"]= df.Date.dt.day
df.head()
```

```python
df["Month_name"]=df.Date.dt.month_name()
df["Day_name"] = df.Date.dt.day_name()
```

```python
df.head()
```

```python
df.info()
```

```python
df['Time_of_day'] = pd.cut(pd.to_datetime(df.Time).dt.hour,
                          bins=[5,12,16,23],
                          labels=['Day','Afternoon','Evening'],
                          include_lowest=True)
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop(["Month"], axis=1, inplace=True)
df.drop(["Day"], axis=1, inplace=True)
```

```python
df.head()
```

```python
df.Year.value_counts()
```

# Summary of Feature Engineering
 1. Some Feature engineering was performed on the dataset for better analysis.
 2. At first we need to change the datatype of the columns Date and Time from object datatyoe to datetime datatype so that the month, day of the week and time of the day could be extractedwhich will provide some useful insights.
 3. Then the month_name and the day_name functions were used to find out the month name and day name.
 4. then the time of day was was categorised as morning, afternoon and evening.
 5. Finally the unnecessary columns were dropped to avoid cluttering of the dataset.
